```python
# Creating a Vending Machine using the Python programming language

# This is where the Menu Will start
# You get to choose whichever snack or drink you want
print(" \033[35m  ---- Welcome to the Vending Machine ---- "),
print(" \033[96m <==========================================>  \033[39m")
print("\nHere's the menu!")
print("What would you like to order?\n")
print("\033[96m===============================================================\033[96m")
print("\033[33mFood		Prices (AED) || Drinks	        Prices (AED)")
print("F1. Ferrero	5            || D1. Pepsi		3.5	")
print("F2. Hersheys	3            || D2. Mogumogu		2.5	")
print("F3. Reeses	4            || D3. Water	        0.5	")
print("F4. Snickers	2.5          || D4. Mountain Dew	3.5	")
print("F5. Maltesers	2.5          || D5. Milk Tea		12	")
print("F6. M&Ms	3.5          || D6. Hot Choco		2.5	\033[39m") 
print("\033[96m===============================================================\033[39m")

# I made this dictionary representing the item code that you will be entering.
# Entering the item code of the item that you choose will make you proceed to the next step
Foods = {"F1" : 'Ferrero', "F2" : 'Hersheys', "F3" : 'Reeses', "F4" : 'Snickers', "F5" : 'Maltesers', "F6" : "M&Ms"}
Drinks = {"D1" : 'Pepsi', "D2" : 'MoguMogu', "D3" : 'Water', "D4" : 'Mountain Dew', "D5" : 'Milk Tea', "D6" : "Hot Choco"}

# These are the list for both the Food and Drinks sections.
# One of these lists will appear after selecting which category you chose.
Foods2 = [{'F1 - Food' : 'Ferrero', 'AED' : 5 }, {'F2 - Food' : 'Hersheys', 'AED' : 3}, {'F3 - Food' : 'Reeses', 'AED' : 4},
          {'F4 - Food' : 'Snickers', 'AED' : 2.5}, {'F5 - Food' : 'Maltesers', 'AED' : 2.5}, {'F6 - Food' : 'M&Ms', 'AED' : 3.5}]
        
Drinks2 = [{'D1 - Drink' : 'Pepsi', 'price' : 3.5}, {'D2 - Drink' : 'MoguMogu', 'price' : 2 }, {'D3 - Drink' : 'Water', 'price' : .5},
          {'D4 - Drink' : 'Mountain Dew', 'price' : 3.5}, {'D5 - Drink' : 'Milk Tea', 'price' : 10.5 }, {'D6 - Drink' : 'Hot Choco', 'price' : 5.5}]

# And these are the prices for both the food and drinks.
fprices = {"F1" : 5, "F2" : 3, "F3" : 4, "F4" : 2.5, "F5" : 2.5, "F6" : 3.5}
dprices = {"D1" : 3.5, "D2" : 2, "D3" : .5, "D4" : 3.5, "D5" : 10.5, "D6" : 5.5}

# From here, you will select an item according to its product code.
# You also have to choose between F or D for food or drinks. (pretty much self explanatory)

while True:
    category = input("\033[4m\nWant Snacks? Press F For the Food Section. Want Drinks? Press D For the Drinks Section: \033[24m").upper()

    if category == 'F':

        for i in Foods2:
            print(F"{i}\n ")

        run = True 
        while run:
            while True:

 # You will then have to choose a product number in order to get your product dispensed.                
                select = (input("What snack would like you to buy?: ")).upper()

 # This Function here selects the item that I have chosen from the Dictionary.
 # If you enter a code that doesn't not make sense or that doesnt exist, it will cause the machine to have an invalid code error.
                if select in Foods:
                    print(f"\nYou have selected: \033[32m{Foods[select]}\033[39m")
                    break
                else:
                    print('\033[91m--invalid code--\033[39m')

 # This will now ask you how many items you want to take from that specific product.
 # This Function will then multiply the prices according to the quantity of the items you ordered.
            while True:
                quantity = int(input(f"\nHow much {Foods[select]} would you like?: "))
                if quantity < 0:
                    print("\033[91mQuantity can not be less than 0.\033[39m")
                elif quantity > 10:
                    print("\033[91mYou can only buy up to 10 items.\033[39m")
                else: 
                    totalFood = float(quantity*fprices[select])
                    print("\033[4m\nYour total is: \033[24m", totalFood)
                    break

 # Here, you will have to enter the amount of money you would want to pay the machine.
 # The function will then deduct the amount of money you entered depending on how much the price of the item/s is.
 # Afterwards, your change will appear.          
            while True:
                money = float(input("Please enter the amount of money you would like to pay: "))
                change = money - totalFood
                if money < totalFood:
                    print("\033[91mNot enough money inserted. Please insert additional funds.\033[39m")
                else:
                    print("\033[4mYour change is: \033[24m" + str(change))
                    break
 # The machine will ask if you want to order again.
 # Saying "Yes" would bring you back to the main menu, so that you can order again.
 # Saying "No" will make the machine give you a receipt then it will end there.
            run2 = True
            while run2:
                repeat = input("\nWould you like to order anything else? ").upper()
                if repeat == "NO":
                    run, run2 = False, False
                    break
                elif repeat == "YES":
                    run2 = False
                else:
                    print("\033[91m--invalid code--\033[39m")
        break
        
 # Everything that I have previously said is also applied in the "drinks section" of the program.        
    elif category == 'D':
        for i in Drinks2:
            print(F"{i}\n ")

        run = True 
        while run:
            while True: 

                select = (input("What drink would like you to buy?: ")).upper()

                if select in Drinks:
                    print(f"\nYou have selected: \033[32m{Drinks[select]}\033[39m")
                    break
                else:
                    print('\033[91m--invalid code--\033[39m')

            while True:
                quantity = int(input(f"\nHow much {Drinks[select]} would you like?: "))
                if quantity < 0:
                    print("\033[91mQuantity can not be less than 0.\033[39m")
                elif quantity > 10:
                    print("\033[91mYou can only buy up to 10 items.\033[39m")
                else: 
                    totalFood = float(quantity*dprices[select])
                    print("Your total is:", totalFood)
                    break
            while True:
                money = float(input("\nPlease enter the amount of money you would like to pay: "))
                change = money - totalFood
                if money < totalFood:
                    print("\033[91Not enough money inserted. Please insert additional funds.\033[39m")
                else:
                    print("Your change is: " + str(change))
                    break
            
            run2 = True
            while run2:
                repeat = input("\nWould you like to order anything else? ").upper()
                if repeat == "NO":
                    run, run2 = False, False
                    break
                elif repeat == "YES":
                    run2 = False
                else:
                    print("\033[91--invalid code--\033[39m")
        break
    
    else:
        print("\033--invalid code--\033[39m")

# The receipt is printed after ordering the item you bought.
# The receipt will show the amount of money you payed, the total price of the item, and the change.
print("\nYour item has been dispensed.")
print("\033[92m\n---Here's your receipt:---")
print(f"you added: {money}")
print(f"total cost: {totalFood}")
print(f"your change: {change}\033[39m")

print("\033[95m<===Thank you, come again!===>\033[39m")

     
     
     

     
```
